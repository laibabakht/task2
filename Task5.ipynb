```python
# Step 1: Install necessary libraries
!pip install -q transformers pandas

# Step 2: Import libraries
from transformers import pipeline
import pandas as pd

# Step 3: Load your CSV file (already uploaded)
df = pd.read_csv("/content/customer_support_tickets.csv")

# Optional: Preview column names
print("✅ Columns loaded:", df.columns.tolist())

# Step 4: Combine 'Ticket Subject' and 'Ticket Description' into one message
df["message"] = df["Ticket Subject"].astype(str) + " - " + df["Ticket Description"].astype(str)

# Step 5: Initialize the zero-shot classifier
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Step 6: Define possible tags
candidate_labels = [
    "billing issue",
    "login problem",
    "technical support",
    "product inquiry",
    "cancellation",
    "account update"
]

# Step 7: Classify each ticket
predicted_tags = []
confidence_scores = []

for ticket in df["message"]:
    result = classifier(ticket, candidate_labels)
    predicted_tags.append(result['labels'][0])
    confidence_scores.append(round(result['scores'][0], 2))

# Step 8: Add predictions to the dataframe
df["predicted_tag"] = predicted_tags
df["confidence"] = confidence_scores

# Step 9: Save to new CSV
output_path = "/content/auto_tagged_tickets.csv"
df.to_csv(output_path, index=False)

print(f"✅ Auto-tagging complete. CSV saved to: {output_path}")
```
